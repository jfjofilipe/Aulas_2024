##Construção de estado de 1 qubit
Nesta secção, vamos criar estados basicos e superposicoes

#### Estado inicial |0>
por padrao, um circuito com 1 qubit inicia no estado |0>, porque representa o estado fundamental do atomo.


```python
#Criando um circuito com 1 qubit no estado |0>
qc = QuantumCircuit(0)  #1 qubit, começa em |0>
qc.draw()            #diagrama do circuito
```

No anterior ja é feito automaticamente. QuantumRegister(1, 'q')


```python
#Criando um circuito com 1 qubit no estado |0>
qReg = QuantumRegister(1,'gubio')
qc = QuantumCircuit(qReg)
qc.draw("mpl")
```

```python
#Estai o vetor de estado resultante do circuito atual
psi = Statevector.from_instruction(qc)
print('Vetor de estado: ',psi)
```

*   Visualização na esfera de bloch

```python
psi0 = Statevector.from_label('0')
display(plot_bloch_multivector(psi0))
```

##Aplicando as operações de qubit
####Aplicando a porta X
A porta X é como NOT quantico e equivale a uma rotação de π(180) em torno do eixo X.




```python
#aplicando X para obter |1>
qc = QuantumCircuit(1)
qc.x(0)                    #aplica X qubit 0
display(qc.draw('mpl'))
```

```python
psi = Statevector.from_instruction(qc)
print('Vetor de estado', psi)
```

```python
#Visualizacao
#|1>
psi = Statevector.from_label('1')
display(plot_bloch_multivector(psi))
```
